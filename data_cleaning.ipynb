```python
import pandas as pd
import numpy as np
data_dim = pd.DataFrame([[2,45,21,np.nan],[56,66,np.nan,np.nan],[8,np.nan,np.nan,np.nan],[np.nan,np.nan,np.nan,np.nan,]])
data_dim

data_dim.dropna()
```

```python
data_dim.dropna(how = 'all')
```

```python
data_dim.dropna(axis =  1, how = 'all')
```

```python
# Drop all columns that have more than 2 NaN values
data_dim.dropna(axis = 1, thresh = 2)
```

```python
# Drop all columns that have more than 2 NaN values
data_dim.dropna(thresh = 2)
```

```python
data_dim
```

```python
# Fill the NaN values with 0
# this one for you to replace NaN to 0 value
data_dim_fill = data_dim.fillna(0)
data_dim_fill
```

```python
# Fill the NaN values with 0
data_dim_fill = data_dim.fillna(2)
print(data_dim_fill)
```

```python
# Pass a dictionary to use differnt values for each column
data_dim_fill = data_dim.fillna({0: 0, 1: 45, 2: 90, 3: 29})
print(data_dim_fill)
```

```python
# Pass method to determine how to fill-up the column - forward here
# this code it will assign value by their self
data_dim_fill = data_dim.fillna(method='ffill')
data_dim_fill
```

```python
# Pass method to determine how to fill-up the column - forward here
# this code it will assign the value at row 3 have limit at 2 column
data_dim_fill = data_dim.fillna(method='ffill', limit = 2)
data_dim_fill
```

```python
# Pass method to determine how to fill-up the row - forward here
# for this code the row at 3 is not assign the any value
data_dim_fill = data_dim.fillna(axis = 1, method='ffill')
data_dim_fill
```
